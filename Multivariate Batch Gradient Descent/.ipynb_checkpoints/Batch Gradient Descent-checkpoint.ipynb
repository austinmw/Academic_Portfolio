```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Reading in the data

```python
# Read in training X (multivariate)
X = pd.read_table("data/X.data", delimiter="   ", header=None,engine='python')
n_features = X.shape[1]
# generate column names
features = []
for i in range(n_features):
    features.append("x_"+str(i+1))
X.columns=features

# Read in training Y (univariate)
Y = pd.read_table("data/Y.data", header=None)

# Check shapes of each
print("X shape: ", X.shape, "\nY shape: ", Y.shape)
```

### Checking for and removing bad values

```python
# Check that length dimensions match
if X.shape[0] == Y.shape[0]:

    # Check for rows with non-Real values
    X_bad_rows = X[~X.applymap(np.isreal).all(1)].index.tolist() # T/F for each row
    Y_bad_rows = Y[~Y.applymap(np.isreal).all(1)].index.tolist()
    print("X non-Real indices: ", X_bad_rows,"\nY non-Real indices: ", Y_bad_rows)

    # Drop rows with non-Real values
    if X_bad_rows == Y_bad_rows: # If non-Real row indices are the same in X and Y:
        X.drop(X_bad_rows, inplace=True)
        Y.drop(X_bad_rows, inplace=True)
    else: # If indices don't match, drop union of bad row lists for both
        bad_rows = list(set(X_bad_rows + Y_bad_rows))
        X.drop(bad_rows, inplace=True)
        Y.drop(bad_rows, inplace=True)

    # recheck size    
    print("X shape: ", X.shape, "\nY shape: ", Y.shape) 

else:
    print("Error: X and Y data have unequal lengths!")
```

### Taking a look at the data

```python
# visualize the data (if low-dimensional)
if X.shape[1] == 1: # if X = [M x 1]
    plt.scatter(X,Y)
elif X.shape[1] == 2: # if x = [M x 2]
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize=(20,20))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X['x_1'], X['x_2'], Y, c='g', marker='o', s=70)
    ax.set_xlabel('sqft.  (x_1)', labelpad=30, FontSize=20)
    ax.set_ylabel('#bedrooms  (x_2)', labelpad=30, FontSize=20)
    ax.set_zlabel('Price  (Y)', labelpad=30, FontSize=20)
    ax.set_title("Training Points for House Data", FontSize=30)
    ax.set_yticks([1,2,3,4,5])
else:
    print("Error: Can't plot %d-dimensional data" % (X.shape[1]+1))
```

### Preprocessing steps to increase gradient descent's efficiency:

- center data
- divide by standard deviations
- append 1's column

```python
# copy data (for easier plotting later since it's a small dataset)
X_copy = X
Y_copy = Y
```

```python
# scale x_1, x_2 by their standard deviations and set means to zeros
sigma = X.std()
mean = X.mean()
X = (X - mean) / sigma

# Add a column of 1's for theta_0 constant term
X['x_0'] = pd.Series(np.ones(X.shape[0]))
X.sort_index(axis=1, inplace=True)

# take a look at the values
X
```

### Calculate cost and evaluate gradient functions

```python
# Calculate cost function
def calc_cost(X,Y,theta):
    m = X.shape[0]
    h = np.dot(X,theta)
    j = (np.dot((h-Y.T),(h-Y.T).T) / (2*m))[0,0]
    return j

# Calculate theta vector update
def eval_gradient(X,Y,theta,alpha):
    m,n = X.shape    
    h = np.dot(X,theta)
    loss = (h-Y.T)   #  1x47
    theta = theta - (np.array((alpha/m) * np.dot(loss,X))[0])
    return theta
```

### Plot cost functions and calculate parameters for different learning rates

```python
# Convert dataframes to numpy matrix objects, array after inspection
X = np.asmatrix(X)
Y = np.array(Y)

# learning rates to try
alphas = [0.01, 0.03, 0.1, 0.3, 1, 1.3, 1.4]

# number of iterations
iterations = 100

# create figure
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

# Loop through alphas 
thetas = [] # list of parameter vectors
for e,a in enumerate(alphas):
    print("alpha: %f" % a)
        
    # Initialize parameters vector
    theta = np.zeros(X.shape[1])    
    
    # Initialize cost function
    J = np.zeros(iterations)

    # Calculate cost function and update parameters with gradient descent step
    for i in range(iterations):
        J[i] = calc_cost(X,Y,theta)   
        theta = eval_gradient(X,Y,theta,a)
    thetas.append(theta) # save each set and use the one with best alpha

             
    # See if learning rate was sufficient by checking if cost function converges nicely
    ax.plot(range(100),J, label=a)

    
    
colormap = plt.cm.nipy_spectral #nipy_spectral, Set1,Paired   
colors = [colormap(i) for i in np.linspace(0, 1,len(ax.lines))]
for i,j in enumerate(ax.lines):
    j.set_color(colors[i])    
    
    
plt.legend()    
plt.ylim([0,70000000000])    
plt.title("Cost (J) as a function of iteration count for different learning rates", FontSize=17)
plt.xlabel("Iteration number", FontSize=15)
plt.ylabel("Cost (J)   [RSS/2m]", FontSize=15)
plt.show()   
```

For the last value of alpha=1.4, gradient descent does not converge at all because this learning rate is too large

### Pick the best looking alpha:

```python
print(alphas)
alpha_choice = float(input("Select a learning rate from below: "))

try:
    ind = alphas.index(alpha_choice)
    theta = thetas[ind]
    print("theta for learning rate %f: " % alpha_choice)
    print(theta)   
except:
    print("\nThe model was not trained on that learning rate!")
```

### Test case:

1650 sqft. house with 3 bedrooms 

Expected prediction is $293,081

```python
# 1650 sqft, 3 bedroom
Test = [1650, 3]

# scale features for prediction and append a 1 for constant term
Test_scaled = [(Test[i]-mean[i])/sigma[i] for i in range(len(Test))]
Test_scaled = [1] + Test_scaled

# Make prediction by taking dot product of Test features and determined parameters
prediction = np.dot(Test_scaled, theta)


print("\n\nPredicted price of a %d sqft. house with %d bedrooms is $%d\n\n" % (Test[0], Test[1], prediction))
```
